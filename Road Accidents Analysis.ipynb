## 3. Road Accidents over the years


```python
df_over_years = df.loc[:, ["YEAR", "TOTAL"]]
df_over_years = df_over_years.groupby("YEAR").sum()['TOTAL']
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df_over_years, color="black",marker='o')
plt.xlabel("Year")
plt.ylabel("Number of Road Accidents ")
plt.title("Number of Road Accidents by Year")
```

## 4. Road Accidents by State

```python
df_state_over_years = df.loc[:, ["TOTAL", "STATE/UT"]]
df_state_over_years = df_state_over_years.groupby("STATE/UT").sum()['TOTAL']
```

```python
plt.figure(figsize=(20,7))
df_state_over_years.plot(kind='bar')
plt.xlabel("STATE")
plt.ylabel("Number of Road Accidents ")
plt.title("Number of Road Accidents by State")
```

```python
def top_n_state_over_year(year):
    df_state = df.loc[:, ['YEAR', 'STATE/UT', 'TOTAL']]
    df_state = df_state[df_state['YEAR'] == year].groupby(['STATE/UT'], as_index=False).sum()\
    .sort_values(by=['YEAR', 'TOTAL'], ascending=[True,False]).head(5)
    plt.figure(figsize=(10,5))
    sns.barplot(x="STATE/UT", y="TOTAL", data=df_state)
    plt.title(f"Top 5 States (Road Accidents in Year {year})")
    plt.xlabel("STATE")
    plt.ylabel("Number of Road Accidents")
    plt.xticks(rotation=90)
    return
```

```python
y = df['TOTAL']
X = df[['0-3 hrs. (Night)','3-6 hrs. (Night)']]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.3,random_state =0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
m=regressor.fit(X_train, Y_train)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score

r2_score(Y_test, y_pred)
```

```python
import numpy as np

MSE=metrics.mean_squared_error(y_pred,Y_test)
RMSE=np.sqrt(MSE)
RMSE
```
